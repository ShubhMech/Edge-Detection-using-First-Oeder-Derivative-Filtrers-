```python
import numpy as np, matplotlib.image as img_, matplotlib.pyplot as plt
```

```python
path=r"D:\Asus\Documents\maxresdefault.jpg"
```

```python
img= img_.imread(path)
```

```python
plt.imshow(img)
```

```python
import cv2
```

```python
img= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
x_sobel=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=7)
```

```python
plt.imshow(x_sobel, cmap='gray')
```

```python
y_sobel=cv2.Sobel(img, cv2.CV_64F, 1,0,ksize=7)
```

```python
plt.imshow(y_sobel, cmap='gray')
```

```python
laplacian= cv2.Laplacian(img, cv2.CV_64F)
```

```python
plt.imshow(laplacian, cmap='gray')
```

```python
threshold_1= 120
```

```python
threshold_2= 200
```

```python
canny_= cv2.Canny(img, threshold_1,threshold_2)
```

```python
plt.imshow(canny_,cmap='gray')
```
